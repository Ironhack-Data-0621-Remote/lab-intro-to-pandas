```python
#Rearrange the columns in the dataframe as needed
df1 = df1.rename(columns={'ST':'State','GENDER':'Gender'})
df1 = df1[['State','Gender','Customer','Education','Income','Customer Lifetime Value','Policy Type','Monthly Premium Auto','Number of Open Complaints','Total Claim Amount','Vehicle Class']]
df1
```

```python
df2 = df2.rename(columns={'ST':'State','GENDER':'Gender'})
df2 = df2[['State','Gender','Customer','Education','Income','Customer Lifetime Value','Policy Type','Monthly Premium Auto','Number of Open Complaints','Total Claim Amount','Vehicle Class']]
df2
```

```python
df3 = df3[['State','Gender','Customer','Education','Income','Customer Lifetime Value','Policy Type','Monthly Premium Auto','Number of Open Complaints','Total Claim Amount','Vehicle Class']]
df3
```

```python
#Concatenate the three DataFrames
data = pd.concat([df1,df2,df3])
data
```

```python
#Which columns are numerical? And categorical
data._get_numeric_data()
```

```python
#Remove the columns education and Number of Open Complaints from the DataFrame
data = data.drop(['Education','Number of Open Complaints'],axis=1)
data
```

```python
#Correct the values in the column Customer Lifetime Value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.
data['Customer Lifetime Value']=pd.to_numeric(data['Customer Lifetime Value'],errors='coerce')
filtered = data[data['Customer Lifetime Value']*100]
filtered
```

```python
#Remove any duplicate rows after check if there are any
data = data.drop_duplicates()
data
```

```python
#Filter out the data for customers who have an income of 0 or less.
Filtered = data[data['Income'] <= 0]
Filtered
```
