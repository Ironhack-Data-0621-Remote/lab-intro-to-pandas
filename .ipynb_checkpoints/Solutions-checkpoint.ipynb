### 3. Standardize header names of the datasets.

Take in count that you have to remove the spaces in the names. 

**hint**: you can use a loop to save time here.

```python
#renaming df1 and df2 columns
df1 = df1.rename(columns={'ST':'State',
                          'GENDER':'Gender'})
df1
```

```python
df2 = df2.rename(columns={'ST':'State',
                          'GENDER':'Gender'})
df2
```

### 4. Rearrange the columns in the dataframe as needed

```python
#Rearrange the columns df2 and df3
df2 = df2[['State', 'Gender', 'Education', 'Customer Lifetime Value', 'Income', 'Monthly Premium Auto', 'Number of Open Complaints', 'Policy Type','Vehicle Class','Total Claim Amount']]
df2
```

```python
df3 = df3[['State', 'Gender', 'Education', 'Customer Lifetime Value', 'Income', 'Monthly Premium Auto', 'Number of Open Complaints', 'Policy Type','Vehicle Class','Total Claim Amount']]
df3
```

### 5. Concatenate the three DataFrames

```python
#Concatenate the three dataframes
data = pd.concat([df1, df2, df3])
data
```

### 6. Which columns are numerical? And categorical?

```python
#Which columns are numerical?
data._get_numeric_data()
```

```python
#Which columns are categorical?
data.select_dtypes('object')
```

### 8. Remove the columns `education` and `Number of Open Complaints` from the DataFrame

```python
#Delete the column education and the number of open complaints from the dataframe
data = data.drop(['Education'], axis=1)
```

```python
data = data.drop(['Number of Open Complaints'], axis=1)
```

```python
data.columns
```

### 9. Correct the values in the column `Customer Lifetime Value`. They are given as a percent, so multiply them by 100 and change dtype to numerical type.

```python
#Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type
data['Customer Lifetime Value'] = pd.to_numeric(data['Customer Lifetime Value'], errors='coerce')
```

```python
data._get_numeric_data()
```

```python
data['Customer Lifetime Value'] = data['Customer Lifetime Value']*100
```

```python
data._get_numeric_data()
```
