### 3. Standardize header names of the datasets.

Take in count that you have to remove the spaces in the names. 

**hint**: you can use a loop to save time here.

```python
df3.columns = df3.columns.str.replace(' ','_').str.lower()
df4.columns = df4.columns.str.replace(' ','_').str.lower()
df5.columns = df5.columns.str.replace(' ','_').str.lower()
df5.rename(columns = {"state" : "st"}, inplace = True)
```

### 4. Rearrange the columns in the dataframe as needed

```python
df4 = df4[['customer', 'st', 'gender', 'education', 'customer_lifetime_value',
       'income', 'monthly_premium_auto', 'number_of_open_complaints',
       'policy_type', 'vehicle_class', 'total_claim_amount']]
df5 = df5 [['customer', 'st', 'gender', 'education', 'customer_lifetime_value',
       'income', 'monthly_premium_auto', 'number_of_open_complaints',
       'policy_type', 'vehicle_class', 'total_claim_amount']]
```

### 5. Concatenate the three DataFrames

```python
frames = [df3, df4, df5]
df_con = pd.concat(frames)
df_con
```

### 8. Remove the columns `education` and `Number of Open Complaints` from the DataFrame

```python
df_con.drop('education', inplace=True, axis=1)
df_con.drop('number_of_open_complaints', inplace=True, axis=1)
```

### 9. Correct the values in the column `Customer Lifetime Value`. They are given as a percent, so multiply them by 100 and change dtype to numerical type.

```python
df_con['customer_lifetime_value'] = df_con['customer_lifetime_value'].str.rstrip('%').astype('float') / 100.0
df_con
```
