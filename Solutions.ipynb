### 3. Standardize header names of the datasets.

Take in count that you have to remove the spaces in the names.

```python
#use for loop to change columns to lower case
cols = []

for i in range(len(df3.columns)):
    cols.append(df3.columns[i].lower())
    
cols
```

```python
cols_2 = []
for i in range(len(df4.columns)):
    cols_2.append(df4.columns[i].lower())
    
cols_2
```

```python
cols_3 = []
for i in range(len(df5.columns)):
    cols_3.append(df5.columns[i].lower())
    
cols_3
```

### 9. Correct the values in the column `Customer Lifetime Value`. They are given as a percent, so multiply them by 100 and change dtype to numerical type.

```python
#removes % sign from column so we can convert it to integer
data['customerlifetimevalue'] = data['customerlifetimevalue'].str.replace("%", '')

```

```python
#converts data in this column to numeric form

data['customerlifetimevalue'] =  pd.to_numeric(data['customerlifetimevalue'], errors='coerce')
```

```python
#mulitplies value of this column by 100
data["customerlifetimevalue"] = 100 * data["customerlifetimevalue"]
```
