### 3. Standardize header names of the datasets.

Take in count that you have to remove the spaces in the names. 

**hint**: you can use a loop to save time here.

```python
#only lowercases for column names in df1
#creating list to store column names with lower cases

cols1 = []

for i in range(len(df1.columns)):
    cols1.append(df1.columns[i].lower())
    
cols1
```

```python
#changing column names in df1
df1.columns = cols1
df1
```

```python
#removing spaces and adding "_" instead for df1
df1.columns = df1.columns.str.replace(' ', '_')
df1
```

```python
#changing column names for df2
df2.columns = df2.columns.str.lower()
df2.columns = df2.columns.str.replace(' ', '_')

df2
```

```python
#changing column names for df3
df3.columns = df3.columns.str.lower()
df3.columns = df3.columns.str.replace(' ', '_')

df3
```
