### 3. Standardize header names of the datasets.

Take in count that you have to remove the spaces in the names. 

**hint**: you can use a loop to save time here.

```python
# rearranging df2 columns
df2 = df2[['Customer', 'ST', 'GENDER', 'Education', 'Customer Lifetime Value',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Vehicle Class', 'Total Claim Amount']]

# rearranging df3 columns
df3 = df3[['Customer', 'State', 'Gender', 'Education', 'Customer Lifetime Value',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Vehicle Class', 'Total Claim Amount']]

# renaming column ST to state in df1 and df2
df1 = df1.rename(columns={'ST':'State'})
df2 = df2.rename(columns={'ST':'State'})

# removing spaces and capital letters
df1.columns = [x.strip(' ').lower() for x in df1.columns]
df2.columns = [x.strip(' ').lower() for x in df2.columns]
df3.columns = [x.strip(' ').lower() for x in df3.columns]
```

```python
# rearranging df2 columns
df2 = df2[['customer', 'state', 'gender', 'education', 'customer lifetime value',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'vehicle class', 'total claim amount']]

# rearranging df3 columns
df3 = df3[['customer', 'state', 'gender', 'education', 'customer lifetime value',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'vehicle class', 'total claim amount']]
```
